```python
def clean_data(file_):
    df = pd.read_csv(file_)
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace("[", "")
    df.columns = df.columns.str.replace("]", "")
    df.columns = df.columns.str.replace(" ", "")
    df = df.drop(columns=["quote_unixtime", "expire_unix"])
    common_cols = [
    'quote_readtime', 'quote_date', 'quote_time_hours', 
    'underlying_last', 'expire_date', 'dte', 
    'strike', 'strike_distance', 'strike_distance_pct'
]

    option_metrics_order = ['delta', 'gamma', 'vega', 'theta', 'rho', 'iv', 'volume', 'last', 'size', 'bid', 'ask']

    call_cols = ['c_delta', 'c_gamma', 'c_vega', 'c_theta', 'c_rho', 'c_iv', 'c_volume', 'c_last', 'c_size', 'c_bid', 'c_ask']
    put_cols  = ['p_delta', 'p_gamma', 'p_vega', 'p_theta', 'p_rho', 'p_iv', 'p_volume', 'p_last', 'p_size', 'p_bid', 'p_ask']

    calls = df[common_cols + call_cols].copy()
    calls = calls.rename(columns={col: col.replace('c_', '') for col in call_cols})
    calls['option_type'] = 'call'

    puts = df[common_cols + put_cols].copy()
    puts = puts.rename(columns={col: col.replace('p_', '') for col in put_cols})
    puts['option_type'] = 'put'

    calls = calls[common_cols + option_metrics_order + ['option_type']]
    puts  = puts[common_cols + option_metrics_order + ['option_type']]
    # for col in calls.columns:
    #     if col in common_cols:
    #         continue
    #     else:
    #         print(col)
    #         calls[col] = calls[col].astype(float)
    #         puts[col] = puts[col].astype(float)
    df = pd.concat([calls, puts], ignore_index=True)
    return df

def save_data(df, filename, folder_path):
    file = filename.strip(".txt").split("_")[2]
    file = 'SPX' + file + ".csv"
    file = folder_path + '/' + file
    print(file)
    df.to_csv(file, index=False)

def open_folder(folder_path):
    return os.listdir(folder_path)
```
